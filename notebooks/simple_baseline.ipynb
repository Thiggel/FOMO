```python
import shutil
import os
# Download the zip file
from google.colab import files

def save_logs_and_checkpoints(logs = True, checkpoints = False):
    # Define the folder to be zipped
    if logs:
        folder_to_download = '/content/FOMO/logs'

        # Create a zip file
        zip_filename = '/content/logs.zip'
        shutil.make_archive('/content/logs', 'zip', folder_to_download)

        files.download(zip_filename)

    if checkpoints:
        folder_to_download = '/content/FOMO/checkpoints'

        # Create a zip file
        zip_filename = '/content/checkpoints.zip'
        shutil.make_archive('/content/checkpoints', 'zip', folder_to_download)

        files.download(zip_filename)
```
